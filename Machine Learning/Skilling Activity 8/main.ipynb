```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Neuron:
    def __init__(self, input_size):
        self.weights = np.random.randn(input_size)
        self.bias = np.random.randn()

    def activation(self, x):
        return 1 / (1 + np.exp(-x))

    def forward(self, x):
        return self.activation(np.dot(x, self.weights) + self.bias)

    def backward(self, x, y, y_hat):
        error = y - y_hat
        self.weights += 0.01 * error * x
        self.bias += 0.01 * error

    def train(self, x, y):
        y_hat = self.forward(x)
        self.backward(x, y, y_hat)
```

```python
Class_A = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 100)
Class_B = np.random.multivariate_normal([3, 3], [[1, 0], [0, 1]], 100)

plt.scatter(Class_A[:, 0], Class_A[:, 1], c='r')
plt.scatter(Class_B[:, 0], Class_B[:, 1], c='b')
```

```python
neuron = Neuron(2)

for i in range(1000):
    for x, y in zip(Class_A, Class_B):
        neuron.train(x, 0)
        neuron.train(y, 1)

x = np.linspace(-2, 4, 100)
y = -(neuron.weights[0] / neuron.weights[1]) * x - (neuron.bias / neuron.weights[1])

plt.plot(x, y, c='g')
plt.scatter(Class_A[:, 0], Class_A[:, 1], c='r')
plt.scatter(Class_B[:, 0], Class_B[:, 1], c='b')
plt.show()
```
