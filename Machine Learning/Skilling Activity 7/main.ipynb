```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Neuron:
    def __init__(self, input_size):
        self.weights = np.random.randn(input_size)
        self.bias = np.random.randn()

    def activation(self, x):
        return x

    def forward(self, x):
        return self.activation(np.dot(x, self.weights) + self.bias)

    def backward(self, x, y, y_hat):
        error = y - y_hat
        self.weights += 0.01 * error * x
        self.bias += 0.01 * error

    def train(self, x, y):
        y_hat = self.forward(x)
        self.backward(x, y, y_hat)
```

```python

X = np.array([[-0.21, 1.49], [0.65, -2.55], [0.33, 0.44], [1.45, 0.14], [0.4, 1.76], [0.38, 0.15], [0.12, 0.76], 
    [-1.45, 2.27], [-1.98, -0.89], [-0.19, 0.05], [-0.85, 0.31], [-0.74, 0.86], [-0.3, -0.39], [2.24, 0.98], 
    [0.41, -0.1], [0.16, -0.35], [1.47, 1.53], [1.2, 1.23], [-0.15, 0.95], [-0.98, 1.87]])

y = np.array([30.71, -61.86, 24.01, 65.21, 53.13, 21.67, 12.29, 18.42, -106.39, -1.08, -17.32, -11.53, -26.81, 
    131.88, -0.85, -8.91, 102.46, 75.81, 20.08, 18.27])

neuron = Neuron(2)

for i in range(1000):
    for j in range(20):
        neuron.train(X[j], y[j])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], y)

x0 = np.linspace(-3, 3, 100)
x1 = np.linspace(-3, 3, 100)
x0, x1 = np.meshgrid(x0, x1)
y_hat = neuron.weights[0] * x0 + neuron.weights[1] * x1 + neuron.bias
ax.plot_surface(x0, x1, y_hat, alpha=0.2)
plt.show()
```
