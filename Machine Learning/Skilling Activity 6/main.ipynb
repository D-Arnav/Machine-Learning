```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Class_A = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 100)
Class_B = np.random.multivariate_normal([3, 3], [[1, 0], [0, 1]], 100)

plt.scatter(Class_A[:, 0], Class_A[:, 1], c='r')
plt.scatter(Class_B[:, 0], Class_B[:, 1], c='b')
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def gradient_descent(X, y, w, alpha, iterations):
    for i in range(iterations):
        w = w - alpha * np.dot(X.T, sigmoid(np.dot(X, w)) - y)
    return w

data = np.concatenate((Class_A, Class_B), axis=0)
data = np.concatenate((np.ones((200, 1)), data), axis=1)
labels = np.concatenate((np.ones((100, 1)), np.zeros((100, 1))), axis=0)

W = np.random.rand(3, 1)
W = gradient_descent(data, labels, W, 0.01, 10000)

x = np.linspace(-4, 6, 100)
y = -(W[0] + W[1]*x)/W[2]

plt.scatter(Class_A[:, 0], Class_A[:, 1], c='r')
plt.scatter(Class_B[:, 0], Class_B[:, 1], c='b')
plt.plot(x, y, c='g')
plt.show()
```
