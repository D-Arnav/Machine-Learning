```python
import numpy as np
from sklearn.neural_network import MLPRegressor
from matplotlib import pyplot as plt
```

```python
x1 = np.random.normal(0, 5, size=100)
x2 = np.random.normal(0, 5, size=100)

def fun(x1, x2):
    return np.cos((x1**2)/4 + (x2**2)/3)

y = np.array([fun(x1[i], x2[i]) for i in range(100)])

model = MLPRegressor(hidden_layer_sizes=(10, 10), activation='logistic', solver='lbfgs', max_iter=1000)
model.fit(np.array([x1, x2]).T, y)

x1_t = np.linspace(0, 5, 100)
x2_t = np.linspace(0, 5, 100)
y_p = np.array([model.predict(np.array([x1_t[i], x2_t[i]]).reshape(1, -1)) for i in range(100)])

plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x1_t, x2_t, y_p)
```
