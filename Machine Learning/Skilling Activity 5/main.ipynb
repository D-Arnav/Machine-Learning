```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Class_A = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 100)
Class_B = np.random.multivariate_normal([3, 3], [[1, 0], [0, 1]], 100)

plt.scatter(Class_A[:, 0], Class_A[:, 1], c='r')
plt.scatter(Class_B[:, 0], Class_B[:, 1], c='b')
```

```python
data = np.concatenate((Class_A, Class_B))
data = np.concatenate((np.ones((200, 1)), data), axis=1) # add bias
labels = np.concatenate((np.ones(100), -np.ones(100)), axis=0)

W = np.dot(np.dot(np.linalg.inv(np.dot(data.T, data)), data.T), labels)

plt.scatter(Class_A[:, 0], Class_A[:, 1], c='r')
plt.scatter(Class_B[:, 0], Class_B[:, 1], c='b')
x = np.array(range(-4, 6))
y = -(W[0] + W[1]*x)/W[2]
plt.plot(x, y, c='k')
plt.show()
```
